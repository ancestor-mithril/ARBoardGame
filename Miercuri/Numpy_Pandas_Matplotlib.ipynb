```python
def get_alchohol_content_by_quality(df):
    alcohol_content_by_quality = []
    for quality in sorted(df['quality'].unique()):
        mask = df['quality'] == quality
        alcohol_content_by_quality.append(df[mask]['alcohol'])
    return alcohol_content_by_quality

plt.figure(figsize=(10, 6))
plt.boxplot(
    get_alchohol_content_by_quality(df),
    labels=sorted(df['quality'].unique())
)
plt.xlabel('Wine Quality')
plt.ylabel('Alcohol Content')
plt.title('Wine Quality vs. Alcohol Content')
plt.show()

# 2 directii
# calitate  mare => cantitate de alcool mai mare
# cantitate de alcool mai mare => calitate mare

# Distinctie intre anomalii statistice si anomalii logice
```

```python
df.head()
```

```python
def get_content_by_quality(df, content):
    alcohol_content_by_quality = []
    for quality in sorted(df['quality'].unique()):
        mask = df['quality'] == quality
        alcohol_content_by_quality.append(df[mask][content])
    return alcohol_content_by_quality

plt.figure(figsize=(10, 6))
plt.boxplot(
    get_content_by_quality(df, 'citric acid'),
    labels=sorted(df['quality'].unique())
)
plt.xlabel('Wine Quality')
plt.ylabel('citric acid')
plt.title('Wine Quality vs. citric acid')
plt.show()
```

```python
df['citric acid'].plot(kind='hist', bins=40, title='citric acid')
plt.show()
```

```python
qualities = df['quality'].to_numpy()

alcohol = df['alcohol'].to_numpy()

rez = np.corrcoef(qualities, alcohol)

print("\t quality, alcool")
print("quality", rez[0])
print("alcool", rez[1])
```

```python

qualities = df['quality'].to_numpy()

sugar = df['residual sugar'].to_numpy()

rez = np.corrcoef(qualities, sugar)

print("\t quality, sugar")
print("quality", rez[0])
print("sugar", rez[1])

```
